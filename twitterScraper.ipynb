```python
import pandas as pd
from ntscraper import Nitter
```

```python
scraper = Nitter()
```

```python
excludeList = ["replies","nativeretweets"]
floodTweets1 = scraper.get_tweets("chennai flood",since="2023-12-02",until="2023-12-06",number=5000,exclude=excludeList,language="en")
```

```python
floodTweets2 = scraper.get_tweets("cyclone michaung",since="2023-12-02",until="2023-12-06",number=2500,exclude=excludeList,language="en")
```

```python
floodTweets = floodTweets1['tweets']+floodTweets2['tweets']
```

```python
floodTweets
```

```python
Tweets = []
for tweet in floodTweets:
    data = [tweet['link'],tweet['user']['name'],tweet['date'],tweet['text'],tweet['stats']['likes'],tweet['stats']['retweets']]
    Tweets.append(data)
```

```python
data = pd.DataFrame(Tweets,columns = ['Link','Name','Date - Time','Tweet','Likes','Retweets'])
data
```

```python
data.to_csv("C:/Users/gowri/Desktop/CAPSTONE/rawDataset.csv", index=False)
```

```python
import re
def remove_non_ascii(text):
    return text.encode("ascii",errors="ignore").decode()

def remove_hashtags(text):
    return re.sub(r'#\w*', '', text)

def remove_mentions(text):
    return re.sub(r'@\w*', '', text)

def remove_urls(text):
    return re.sub(r'https?://\S+|www\.\S+', '', text)
```

```python
newTweet = []
for tweet in data['Tweet']:
    tweet = tweet.lower()
    tweet = remove_non_ascii(tweet)
    tweet = remove_hashtags(tweet)
    tweet = remove_mentions(tweet)
    tweet = remove_urls(tweet)
    newTweet.append(tweet)
    
data['Tweet'] = newTweet
data.to_csv("C:/Users/gowri/Desktop/CAPSTONE/Dataset.csv", index=False)
```
